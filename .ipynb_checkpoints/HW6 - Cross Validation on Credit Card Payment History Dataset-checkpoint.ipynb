```python
#Import select packages
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score

import pandas as pd
import numpy as np
import statistics 
```

```python
#Import credit card payment history dataset
df = pd.read_csv(r"C:\Users\dmoha\Downloads\ccdefault.csv")
df = df.dropna()

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df.head()
```

```python
#Summary statistics on the creditcard dataset
df.describe()
```

```python
X = df.iloc[:, 1:-1]
y = df.iloc[:, -1]
```

#### Part 1: Random test train splits
Run in-sample and out-of-sample accuracy scores for 10 different samples by changing random_state from 1 to 10 in sequence. 
Display the individual scores, then calculate the mean and standard deviation on the set of scores.  Report in a table format.


```python
in_sample_list = []
out_sample_list = []

for i in range(11):
    dt = DecisionTreeClassifier(max_depth = 2, random_state = i)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = i)
    dt.fit(X_train, y_train)
    
    y_pred_test = dt.predict(X_test)
    y_pred_train = dt.predict(X_train)
    
    inSampleAccuracy = accuracy_score(y_train, y_pred_train)
    outSampleAccuracy = accuracy_score(y_test, y_pred_test)
        
    in_sample_list.append(inSampleAccuracy)
    out_sample_list.append(outSampleAccuracy)
    
accuracyScore_DF = pd.DataFrame(
    {
        'Random State': range(11),
        'in-sample accuracy': in_sample_list,
        'out-of-sample accuracy': out_sample_list
    })

accuracyScore_DF.head(10)
```

```python
accuracyScore_DF[['in-sample accuracy', 'out-of-sample accuracy']].describe()
```

#### Part 2: Cross validation
Now rerun your model using cross_val_scores with k-fold CV (k=10). Report the individual fold accuracy scores, the mean CV score and the standard deviation of the fold scores. Now run the out-of-sample accuracy score. Report in a table format.


```python
dtCV = DecisionTreeClassifier(max_depth = 2, random_state = 42)

cv_scores_train = cross_val_score(dtCV, X_train, y_train, cv=10)
cv_scores_test = cross_val_score(dtCV, X_test, y_test, cv=10)

CV_accuracyScore_DF = pd.DataFrame(
    {
        'CV Fold': range(10),
        'in-sample CV scores': cv_scores_train,
        'out-of-sample CV scores': cv_scores_test
    })

CV_accuracyScore_DF.head(10)
```

```python
CV_accuracyScore_DF[['in-sample CV scores', 'out-of-sample CV scores']].describe()
```
